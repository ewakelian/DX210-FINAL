```python
data = pd.read_csv('iris.csv')
data
```

```python
df = pd.DataFrame(data)
df
```

```python
data_histogram1 = df["sepal_length"]
m = [m for m in range (len(data_histogram1))]
plt.bar(m,data_histogram1.values)
plt.title("sepal length Raw Data")
plt.show()
```

```python
data_histogram2 = df["sepal_width"]
m = [m for m in range (len(data_histogram1))]
plt.bar(m,data_histogram1.values)
plt.title("sepal width Raw Data")
plt.show()
```

```python
plt.hist(data_histogram2.values, cumulative = True)
plt.title("sepal width Cumulative Data")
plt.show()
```

```python
data_histogram3 = df["petal_length"]
m = [m for m in range (len(data_histogram3))]
plt.bar(m,data_histogram3.values)
plt.title("petal length Raw Data")
plt.show()
```

```python
data_histogram4 = df["petal_width"]
m = [m for m in range (len(data_histogram4))]
plt.bar(m,data_histogram4.values)
plt.title("petal width Raw Data")
plt.show()
```
