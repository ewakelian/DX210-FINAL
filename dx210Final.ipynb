```python
x = x[~np.isnan(y)];
y = y[~np.isnan(y)];
```

```python
from sklearn.datasets import make_classification
from collections import Counter
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
```

```python
x,y = make_classification(n_classes=2, n_redundant=0,weights=[0.1,0.9],n_features=2, n_samples=500,random_state=4)
```

```python
print(Counter(y).values())
print(Counter(y))
```

```python
sm = SMOTE()
x_resampled,y_resampled = sm.fit_resample(x,y)
print(Counter(y_resampled).values())
```

```python
print(Counter(y_resampled).values())
```

```python
plt.scatter(x_resampled[:,0], x_resampled[:,1], color='blue' , linewidths=0.1, label="Upsampling")
plt.scatter(x[:,0], x[:,1], color='orange', linewidths=0.1, label="original")

plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(x)
df['class'] = y
df
```

```python
print(Counter(df['class']))
```

```python
from sklearn.utils import resample
```

```python
df_minority = df[df.iloc[:,2] == 0]
df_majority = df[df.iloc[:,2] == 1]
```

```python
df_majority_downsampled = resample(df_majority,
                                   n_samples=53,
                                   random_state=123)
```

```python
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
df_downsampled
```

```python
print(Counter(df_downsampled['class']).values())
```

```python
import numpy as np
colors = np.array(['blue','red'])

plt.subplot(1,2,1)
plt.scatter(df.iloc[:,0],df.iloc[:,1], color=colors[df['class']], linewidths=0.1)
plt.title('Original data')
plt.subplot(1,2,2)
plt.scatter(df_downsampled.iloc[:,0],df_downsampled.iloc[:,1],color=colors[df_downsampled['class']], linewidths=0.1)
plt.title('Downsampling')
plt.show()
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data210/bank_call_center.csv')
data
```

```python
data = data.drop(columns=['Customer_ID'])
data
```

```python
df = pd.DataFrame(data)
df
```

```python
X = data.values
print(X)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_x = LabelEncoder()
for i in range(1,8):
  X[:,i] = labelencoder_x.fit_transform(X[:,i])
```

```python
X = scale(X)
X
```

```python
wcss = []
for i in range(1, 8):
    kmeans = KMeans(n_clusters=i, max_iter=300, n_init=1, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 8), wcss,  'o-')
plt.xticks(range(1,8))
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
!pip install kneed
from kneed import KneeLocator
```

```python
kneedle = KneeLocator(range(1, 8), wcss,
                      curve='convex',
                      direction='decreasing')
kneedle.elbow
```

```python
kneedle.plot_knee()
```

```python
km = KMeans(n_clusters=2, random_state=0, max_iter=300, init='random')
km.fit_predict(X)
```

```python
km.cluster_centers_
```

```python
df['labels'] = km.labels_
```

```python
df0 = df[df.labels == 0]
df1 = df[df.labels == 1]
```
